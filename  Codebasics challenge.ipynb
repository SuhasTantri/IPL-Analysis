# Top 10 Batsmen on the basis of total runs scored

```python
df_batting.groupby(['batsmanName']).sum().sort_values(by='runs',ascending=False).head(10)
```

```python
df_bat_match = df_batting.merge(df_match,on = 'match_id',how='inner')
df_bat_match.head()
```

```python
df_bat_match['year'] = pd.to_datetime(df_bat_match['matchDate']).dt.year
df_bat_match['year'].replace(2029,2023,inplace=True)
```

```python
df_filter = pd.DataFrame(df_bat_match.groupby(['year','batsmanName']).sum())
df_filter
```

```python
df_filter = df_filter[df_filter['balls']>60].reset_index()
df_filter
```

```python
df_2 = pd.DataFrame(df_filter['batsmanName'].value_counts().reset_index()).reset_index()
df_3_seasons=df_2[df_2['batsmanName']>2]
df_3_seasons
```

```python
df_batting1 = df_batting.replace({'out': 1,'not_out':0})
df_batting1
```

```python
df_bat_average = df_batting1.groupby(['batsmanName']).sum().reset_index()
```

```python
df_bat_average['average'] = df_bat_average['runs']/df_bat_average['out/not_out']
```

```python
df_bat_average = df_bat_average.sort_values(by = 'average',ascending=False)
df_bat_average.head()
```

# Top 10 batsmen based on Batting Average over past 3 seasons (atleast 60 balls faced over each of past 3 seasons)

```python
df_bat_avg = df_bat_average[df_bat_average['batsmanName'].isin(df_3_seasons['index'])]
df_bat_avg[['batsmanName','average']].head(10)
```

# Top 10 Batsmen based on Strike Rate over past 3 seasons (min 60 balls faced over each of past 3 seasons)

```python
df_bat_avg['strike_rate'] = df_bat_avg['runs']*100/df_bat_avg['balls']
df_strk_rate = df_bat_avg.sort_values(by='strike_rate',ascending=False)
df_strk_rate[['batsmanName','strike_rate']].head(10)
```

# Top 10 bowlers based on most wickets taken over past 3 years

```python
df_bowling_wickets = df_bowling.groupby(['bowlerName']).sum().sort_values(by='wickets',ascending=False).reset_index()
```

```python
df_bowling_wickets[['bowlerName','wickets']].sort_values(by='wickets',ascending=False).head(10)
```

```python
df_bowling.head()
```

```python
# Joining two dataframes 
df_bowl_match = df_bowling.merge(df_match,on = 'match_id',how='inner')
df_bowl_match.head()
```

```python
df_bowl_match['year'] = pd.to_datetime(df_bowl_match['matchDate']).dt.year
df_bowl_match['year'].replace(2029,2023,inplace=True)
```

```python
df_bowl_match['year'].value_counts()
```

```python
df_bowl_filter = df_bowl_match.groupby(['year','bowlerName']).sum().reset_index()
```

```python
# Identifying bowlers who have bowled atleast 10 overs 
df_bowl_filter = df_bowl_filter[df_bowl_filter['overs']>=10]
df_bowl_filter_2 = df_bowl_filter['bowlerName'].value_counts().reset_index()
# Identifiying bowlers who have bowled over past 3 seasons
df_bowl_2 = df_bowl_filter_2[df_bowl_filter_2['bowlerName']>2]
df_bowl_2
```

# Top 10 bowlers based on bowling average over past 3 seasons (atleast 60 balls over each of past 3 years)

```python
# Filtering bowlers who have bowled atleast 10 over each of past 3 seasons and calculating average and finding top 10 bowlers based on average
df_bowl_average = df_bowl_match[df_bowl_match['bowlerName'].isin(df_bowl_2['index'])]
df_bowl_average_1 = df_bowl_average.groupby(['bowlerName']).sum().reset_index()
df_bowl_average_1['average'] = round(df_bowl_average_1['runs']/df_bowl_average_1['wickets'],2)
df_10_bowl_average = df_bowl_average_1[['bowlerName','average']].sort_values(by='average',ascending=True).reset_index().head(10)
df_10_bowl_average
```

# Top 10 bowlers based on economies over past 3 seasons (atleast 60 balls over each of past 3 years)

```python
# Filterig bowlers who have bowled atleast 60 balls over past 3 seasons and grouping the bowler names and average of economies
df_economy = df_bowl_average[df_bowl_average['bowlerName'].isin(df_bowl_2['index'])]
df_10_economy = df_economy[['bowlerName','economy']].groupby(['bowlerName']).mean().reset_index().sort_values(by='economy',ascending=True)
df_10_economy.head(10)
```

# Top 5 batsmen based on past 3 years boundary percentage

```python
df_boundary = df_batting.groupby('batsmanName').sum().reset_index()
df_boundary['boundary%'] = round((df_boundary['4s']+df_boundary['6s'])*100/df_boundary['balls'],2)
df_boundary.sort_values(by='boundary%',ascending=False).head()
```

# Top 5 bowlers based on dot ball % who have bowled atleast 40 overs

```python
#
df_bowling['balls'] = (np.floor(df_bowling['overs'])*6)+((df_bowling['overs']-np.floor(df_bowling['overs']))*10)
df_bowling = df_bowling.groupby('bowlerName').sum().reset_index()
df_bowling = df_bowling[df_bowling['overs']>=40]
df_bowling['dotball%'] = round(df_bowling['0s']*100/df_bowling['balls'],2)
df_bowling.sort_values(by='dotball%',ascending=False)
```

```python
df_bowling[['bowlerName','dotball%']].sort_values(by='dotball%',ascending=False).head()
```
